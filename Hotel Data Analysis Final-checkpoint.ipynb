```python
# Changing the datatype of reservation_status_date column from object type to date

df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
df['reservation_status_date'].dtype
```

```python
# dropping the agent and company columns

df.drop(['company','agent'],axis=1,inplace=True)

# dropping the rows with missing values in the country column

df.dropna(inplace=True)
```

```python
# Null values in each column after data cleaning

df.isnull().sum()
```

```python
# Summary statistics of numerical columns

df.describe()
```

```python
# Outlier in the adr(average daily rate) column

plt.figure(figsize=(5,5))
df['adr'].plot(kind='box')
```

```python
# Removing the outlier present in the 'adr' column

df=df[df['adr']<5000]
df=df[df['adr']>0]
```

```python
df.describe()
```

# Data Analysis and Visualizations

```python
# Checking the percentage of cancelled and non-cancelled reservations 
# 0 = not cancelled
# 1 = cancelled

cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc*100)
```

```python
plt.figure(figsize=(5,4))
plt.bar(['Not canceled','Canceled'],df['is_canceled'].value_counts(),width=0.6)
plt.title('Reservation Status Count')
```

```python
# Which hotel has more cancellation rate
# 0 = not cancelled
# 1 = cancelled

plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title('Reservation Status in Different Hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not cancelled','Cancelled'])
plt.show()
```

```python
# Percentage of non cancelled and cancelled reservations
# Resort Hotel Vs. City Hotel
# 0 = not cancelled
# 1 = cancelled

resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel_cancel=resort_hotel['is_canceled'].value_counts(normalize=True)
print('Resort hotel percentages')
print(resort_hotel_cancel*100)
print('-'*50)
city_hotel=df[df['hotel']=='City Hotel']
city_hotel_cancel=city_hotel['is_canceled'].value_counts(normalize=True)
print('City hotel percentages')
print(city_hotel_cancel*100)
```

```python
# Average daily price of different hotels

resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
print('Average daily price of resort hotel')
print(resort_hotel)
print('-'*60)
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
print('Average daily price of city hotel')
print(city_hotel)
```

```python
# Plotting the price per day difference in two different hotel types

plt.figure(figsize=(20,8))
plt.title('Average Daily Rate of City Vs. Resort Hotel (2014 - 2017)',fontsize=30)
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
# Cancellations per month

df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(20,10))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.title('Reservation status per month',fontsize=30)
plt.xlabel('Months')
plt.ylabel('Number of reservations')
plt.legend(['Not cancelled', 'Cancelled'],fontsize=20)
plt.show()
```

```python
# Average daily rate of hotel rooms by month for cancellations

plt.figure(figsize=(15,8))
plt.title('ADR per month',fontsize=30)
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.ylabel('ADR')
plt.xlabel('Months')
plt.show()
```

```python
# Countries with highest cancellation rate

cancelled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation cancellation')
plt.pie(top_10_country,labels=top_10_country.index,autopct='%.2f')
plt.show()

# PRT (portugal faces the highest cancelation)
```

```python
# Market segmentation 
# How does the customers make bookings for the hotels

df['market_segment'].value_counts()
```

```python
# Percent market segmentation

df['market_segment'].value_counts(normalize=True)
```

```python
# Percent cancellation by each market segemnt

cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
# Highest percentage of reservation cancellations is coming from the Online travel agents (TA)
# who also holds the major market segment in hotel bookings  
```

```python
# Comparison of Average Daily Rate of cancelled Vs. Non-canceled reservations
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled.groupby('reservation_status_date')[['adr']].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(25,10))
plt.title('Average Daily Rate (Cancelled Vs Non-cancelled reservations)',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='Cancelled')
plt.legend(fontsize=20)
plt.plot()
```

```python
# Filtering the data from 2016-01 to 2017-09 to get better idea 

cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
plt.figure(figsize=(25,10))
plt.title('Average Daily Rate (Cancelled Vs Non-cancelled reservations)',fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='Not Cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='Cancelled')
plt.legend(fontsize=20)
plt.plot()
```
